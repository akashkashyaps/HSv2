```python
import re
# Define a function to extract context and answer using regular expressions
def extract_context_answer(text):
    match = re.search(r'CONTEXT:(.*?)QUESTION:(.*?)end_header_id(.*)$', text, re.DOTALL)
    if match:
        context = match.group(1).strip()
        answer = match.group(3).strip().replace("\n", " ").replace("\r", "").replace("|>", "")
        return context, answer
    else:
        return None, None
```

```python
results_set[['context', 'answer']] = results_set['Answer'].apply(lambda x: pd.Series(extract_context_answer(x)))
results_set
```

```python
results_set = results_set.drop(columns=['Answer'])
results_set
```

```python
results_set = results_set.rename(columns={'Question': 'question'})
results_set
```

```python
evaluation_set_LLaMa3 = pd.merge(results_set, test_set, on='question')
evaluation_set_LLaMa3
```

```python
evaluation_set_LLaMa3.drop_duplicates(inplace=True)
evaluation_set_LLaMa3.dropna(inplace=True)
evaluation_set_LLaMa3
```
