```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("evaluation_set_Advanced.csv")
df
```

```python
df.rename(columns={'contexts':'RAGAS_Contexts','context':'Advanced_contexts'},inplace=True)
```

```python
import re
import nltk
from bert_score import score
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('english'))
def clean_text(text):
    # Decode common encoding issues
    text = text.replace('â€“', '-').replace('â€™', "'").replace('â€œ', '"').replace('â€', '"')
    # Convert to lowercase
    text = text.lower()
    # Remove punctuation and special characters
    text = re.sub(r'[^\w\s]', '', text)
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    # Remove stopwords
    text = " ".join([word for word in text.split() if word not in stop_words])

    return text
```

```python
# Apply text cleaning to both columns
df['RAGAS_Contexts'] = df['RAGAS_Contexts'].apply(clean_text)
df['Advanced_contexts'] = df['Advanced_contexts'].apply(clean_text)
df
```

```python
# Function to calculate BERTScore similarity
def calculate_bertscore_similarity(row):
    P, R, F1 = score([row['RAGAS_Contexts']], [row['Advanced_contexts']], lang='en', model_type='bert-base-uncased')
    return F1.item()  # F1 score is usually the best single metric to use

# Apply the function to each row
df['contexts_bertscore_similarity'] = df.apply(calculate_bertscore_similarity, axis=1)
```
