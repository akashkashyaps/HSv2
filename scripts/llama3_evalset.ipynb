```python
import re
# Define a function to extract context and answer using regular expressions
def extract_context_answer(text):
    match = re.search(r'CONTEXT:(.*?)QUESTION:(.*?)end_header_id(.*)$', text, re.DOTALL)
    if match:
        context = match.group(1).strip()
        answer = match.group(3).strip().replace("\n", " ").replace("\r", "").replace("|>", "")
        return context, answer
    else:
        return None, None
```

```python
results_set[['context', 'answer']] = results_set['Answer'].apply(lambda x: pd.Series(extract_context_answer(x)))
results_set
```

```python
llama_inference = pd.read_csv("Llama_time.csv")
llama_inference
```

```python
import re
# Define a function to extract context and answer using regular expressions
def extract_context_answer(text):
    match = re.search(r'CONTEXT:(.*?)QUESTION:(.*?)end_header_id(.*)$', text, re.DOTALL)
    if match:
        context = match.group(1).strip()
        answer = match.group(3).strip().replace("\n", " ").replace("\r", "").replace("|>", "")
        return context, answer
    else:
        return None, None
```

```python
llama_inference[['context', 'answer']] = llama_inference['Answer'].apply(lambda x: pd.Series(extract_context_answer(x)))
llama_inference
```

```python
import nltk
from nltk import word_tokenize
nltk.download('punkt')
```

```python
# Function to count words in each row of the specified column
def count_words(answer):
    words = word_tokenize(answer)
    return len(words)
```

```python
llama_inference['total_tokens'] = llama_inference['answer'].apply(count_words)
llama_inference
```

```python
llama_inference["tokens_per_second"] = llama_inference["total_tokens"] / llama_inference["Time_Taken"]
llama_inference
```

```python
llama_inference.drop(['Answer', 'context'], axis=1, inplace=True)
llama_inference
```

```python
llama_inference.to_csv('llama_inference.csv', index=False)
```
