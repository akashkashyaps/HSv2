```python
import pandas as pd
import numpy as np
```

```python
results_set = pd.read_csv('results_mistral.csv')
test_set = pd.read_csv('testset7-cleaned-JB-FFT.csv')
```

```python
import re
# Define a function to extract context and answer using regular expressions
def extract_context_answer(text):
    match = re.search(r'CONTEXT:(.*?)QUESTION:(.*?)INST(.*)$', text, re.DOTALL)
    if match:
        context = match.group(1).strip()
        answer = match.group(3).strip().replace("\n", " ").replace("\r", "").replace("[/", "").replace("]", "")
        return context, answer
    else:
        return None, None
```

```python
results_set[['context', 'answer']] = results_set['Answer'].apply(lambda x: pd.Series(extract_context_answer(x)))
results_set
```

```python
results_set = results_set.drop(columns=['Answer'])
results_set
```

```python
results_set = results_set.rename(columns={'Question': 'question'})
results_set
```

```python
evaluation_set_Mistral7B = pd.merge(results_set, test_set, on='question')
evaluation_set_Mistral7B
```

```python
evaluation_set_Mistral7B.drop_duplicates(inplace=True)
evaluation_set_Mistral7B.dropna(inplace=True)
evaluation_set_Mistral7B
```

```python
evaluation_set_Mistral7B.to_csv('evaluation_set_Mistral7B.csv', index=False)
```

```python
mistral_inference = pd.read_csv('Mistral_time.csv')
mistral_inference
```

```python
import re
# Define a function to extract context and answer using regular expressions
def extract_context_answer(text):
    match = re.search(r'CONTEXT:(.*?)QUESTION:(.*?)INST(.*)$', text, re.DOTALL)
    if match:
        context = match.group(1).strip()
        answer = match.group(3).strip().replace("\n", " ").replace("\r", "").replace("[/", "").replace("]", "")
        return context, answer
    else:
        return None, None
```

```python
mistral_inference[['context', 'answer']] = mistral_inference['Answer'].apply(lambda x: pd.Series(extract_context_answer(x)))
mistral_inference
```

```python
import nltk
from nltk import word_tokenize
nltk.download('punkt')
```

```python
# Function to count words in each row of the specified column
def count_words(answer):
    words = word_tokenize(answer)
    return len(words)
```

```python
mistral_inference['total_tokens'] = mistral_inference['answer'].apply(count_words)
mistral_inference
```

```python
mistral_inference["tokens_per_second"] = mistral_inference["total_tokens"] / mistral_inference["Time_Taken"]
mistral_inference
```

```python
mistral_inference.drop(['Answer', 'context'], axis=1, inplace=True)
mistral_inference
```

```python
mistral_inference.to_csv('mistral_inference.csv', index=False)
```
