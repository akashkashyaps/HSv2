```python
import re
pattern = r"CONTEXT:(.*?)QUESTION:(.*?)Helpful Answer:(.*)"
```

```python
# Define a function to apply the regular expression pattern and extract context and answer
def extract_context_answer(text):
    match = re.search(pattern, text, re.DOTALL)
    if match:
        context = match.group(1).strip()
        answer = match.group(3).strip()
        return context, answer
    else:
        return None, None
```

```python
results_set[['context', 'answer']] = results_set['Answer'].apply(lambda x: pd.Series(extract_context_answer(x)))
results_set
```

```python
phi_inference = pd.read_csv('Phi_time.csv')
phi_inference
```

```python
import re
pattern = r"CONTEXT:(.*?)QUESTION:(.*?)Helpful Answer:(.*)"
```

```python
# Define a function to apply the regular expression pattern and extract context and answer
def extract_context_answer(text):
    match = re.search(pattern, text, re.DOTALL)
    if match:
        context = match.group(1).strip()
        answer = match.group(3).strip()
        return context, answer
    else:
        return None, None
```

```python
phi_inference[['context', 'answer']] = phi_inference['Answer'].apply(lambda x: pd.Series(extract_context_answer(x)))
phi_inference
```

```python
import nltk
from nltk import word_tokenize
nltk.download('punkt')
```

```python
# Function to count words in each row of the specified column
def count_words(answer):
    words = word_tokenize(answer)
    return len(words)
```

```python
phi_inference['total_tokens'] = phi_inference['answer'].apply(count_words)
phi_inference
```

```python
phi_inference["tokens_per_second"] = phi_inference["total_tokens"] / phi_inference["Time_Taken"]
phi_inference
```

```python
phi_inference.drop(['Answer', 'context'], axis=1, inplace=True)
phi_inference
```

```python
phi_inference.to_csv('phi_inference.csv', index=False)
```
